```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# The data set is all set to go and does not contain any missing value even though there is NaN value in the data set but that just indicated a person did not has sleep disorder
# and we don't have to worry about it since if a person don't have sleep disorder then we don't have to investigate to get treatment for them


df = pd.read_csv('EASminiprojectDataset.csv') # Load data set

avg_sleep = df.groupby('Occupation')['Sleep Duration'].mean().sort_values(ascending=False).reset_index() # Get the average Sleep duration
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_sleep, x='Sleep Duration', y='Occupation', palette='coolwarm')
plt.title('Average Sleep Duration by Occupation')
plt.xlabel('Average Sleep Duration (hrs)')
plt.ylabel('Occupation')
plt.tight_layout()
plt.show()
```

```python
# Filter for BMI Categories of interest
bmi_data = df[df['BMI Category'].isin(['Overweight', 'Normal'])]

# First plot: Average Daily Steps
avg_steps = bmi_data.groupby('BMI Category')['Daily Steps'].mean().reset_index()

# Second plot: Count of Sleep Disorders per BMI category
sleep_counts = bmi_data.groupby(['BMI Category', 'Sleep Disorder']).size().unstack(fill_value=0)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Bar chart for average daily steps
axes[0].bar(avg_steps['BMI Category'], avg_steps['Daily Steps'], color=['#66c2a5', '#fc8d62'])
axes[0].set_title('Average Daily Steps\n(Obese vs Normal Individuals)')
axes[0].set_ylabel('Average Daily Steps')
axes[0].set_xlabel('BMI Category')

# Bar chart for sleep disorder counts
sleep_counts.plot(kind='bar', stacked=True, ax=axes[1], colormap='coolwarm')
axes[1].set_title('Count of Sleep Disorders\nby BMI Category')
axes[1].set_ylabel('Count')
axes[1].set_xlabel('BMI Category')
axes[1].legend(title='Sleep Disorder')


plt.tight_layout()
plt.show()
```

```python
disorder_counts = df.groupby(['Sleep Disorder', 'Occupation']).size().reset_index(name='Count') # Get the count for Sleep Disorder for each occupation
disorder_counts = disorder_counts.sort_values('Count', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=disorder_counts, x='Count', y='Occupation', hue='Sleep Disorder', dodge=True)
plt.title('Sleep Disorders by Occupation')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Stress Level', y='Quality of Sleep', palette='Set3')
plt.title('Quality of Sleep by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Quality of Sleep')
plt.tight_layout()
plt.show()
```

```python
avg_stress = df.groupby('Occupation')['Stress Level'].mean().sort_values(ascending=False).reset_index() # Average stress for each occupation

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_stress, x='Stress Level', y='Occupation', palette='flare')
plt.title('Average Stress Level by Occupation')
plt.xlabel('Average Stress Level')
plt.ylabel('Occupation')
plt.tight_layout()
plt.show()
```
